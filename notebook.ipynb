## Import libraries

In this notebook, I use TensorFlow/Keras to build a model for handwritten digit classification.  
I also use NumPy for data handling and `plot_model` to create the model diagram.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.utils import plot_model
```

## Load the dataset

Here I load the MNIST dataset.  
It contains handwritten digit images from 0 to 9.  
The test set has 10,000 images, which will be used for the final score.

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

print("Training images:", X_train.shape)
print("Test images:", X_test.shape)
```

## Prepare the data

Before training, I normalize the pixel values from 0–255 to 0–1.  
I also add one channel dimension because the CNN model expects image data in that format.

```python
# Normalize pixel values
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0

# Add channel dimension for CNN
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)

print("Training shape after preprocessing:", X_train.shape)
print("Test shape after preprocessing:", X_test.shape)
```

## Build the model

In this step, I create a CNN model.  
CNN is a good choice for image classification because it can learn useful patterns from images, such as edges and shapes.

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(28, 28, 1)),

    layers.Conv2D(16, (3, 3), activation="relu", padding="same"),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(32, (3, 3), activation="relu", padding="same"),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation="relu", padding="same"),

    layers.GlobalAveragePooling2D(),

    layers.Dense(32, activation="relu"),
    layers.Dense(10, activation="softmax")
])
```

## Compile the model

Now I compile the model.  
I use the Adam optimizer, sparse categorical crossentropy as the loss function, and accuracy as the evaluation metric.

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

## Train the model

Here I train the model using the training dataset.  
I also use 20% of the training data as validation data to monitor performance during training.

```python
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2
)
```

## Save the model as a `.keras` file

After training, I save the model as a `.keras` file.  
This file stores the model structure and trained weights.

```python
model.save("mnist_model.keras")
print("Model saved successfully!")
```

## Evaluate the model on the test set

Now I test the model on the 10,000-image test set.  
I calculate the number of correct predictions, because this is used in the reward formula.

```python
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)

correct_predictions = int(round(test_acc * len(X_test)))

print("Test accuracy:", test_acc)
print("Correct predictions:", correct_predictions)
```

## Calculate the final score

The final score is based on two things:

1. Reward for correct classifications  
2. Penalty for the total number of model parameters

The formula is:

**Net Score = Reward − Total Parameters**

```python
total_params = model.count_params()

if correct_predictions <= 5000:
    reward = correct_predictions * 100
elif correct_predictions <= 6000:
    reward = (5000 * 100) + ((correct_predictions - 5000) * 200)
else:
    reward = (5000 * 100) + (1000 * 200) + ((correct_predictions - 6000) * 1000)

net_score = reward - total_params

print("Total parameters:", total_params)
print("Reward:", reward)
print("Net Score:", net_score)
```

##  example predictions

Finally, I display a few test images together with their predicted labels to visually check the model performance.

```python
import matplotlib.pyplot as plt

predictions = model.predict(X_test, verbose=0)
predicted_labels = np.argmax(predictions, axis=1)

plt.figure(figsize=(10, 4))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_test[i].squeeze(), cmap="gray")
    plt.title(f"True: {y_test[i]}, Pred: {predicted_labels[i]}")
    plt.axis("off")
plt.tight_layout()
plt.show()
```
